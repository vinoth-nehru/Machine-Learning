```python
specs=[[400,1],[1000,2],[200,6],[150,4],[0,0]]
lables=['super-car','audi-r8','suv-car','family-car','Hey its not a car']
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(specs,lables)
```

```python
model.predict([[1,1],[2000,2]])
```

```python
import matplotlib.pyplot as plt
plt.imshow(specs, interpolation='none')
plt.show()
```

```python
plt.plot(lables,specs)
plt.show()
```

```python
power=[i[0] for i in specs]
seats=[i[1] for i in specs]
plt.bar(seats,power)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(seats,power)

for i, txt in enumerate(lables):
    ax.annotate(txt, (seats[i], power[i]))
```
